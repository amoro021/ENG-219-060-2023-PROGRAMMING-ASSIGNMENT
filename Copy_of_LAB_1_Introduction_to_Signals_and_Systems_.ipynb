## Types of Signals

We represent signals using mathematical functions. **A function** maps elements in the domain set to the elements in the co-domain/ range set, thats is,

$$ f: \mathcal{X} \to \mathcal{Y} \\ \qquad \quad f(x) \mapsto y$$  where $\mathcal{X}$ is the domain and $\mathcal{Y}$ is the range or the co-domain set of elements. Such functions are also simple expressed using the equation:
$$ y = f(x) \quad \text{where} \quad x \in \mathcal{X} \quad \text{amd} \quad y \in \mathcal{Y}.$$

In this notebook, we shall deal with functions of single independent variables, which usually represent time, denoted by $t$ or $n$ for continuous time (C.T) and discrete time (D.T.) signals, respectively. Of course we shall deviate a little bit and explore multi-dimensional signals, random signals (stochastic processes) and graph signals which might require us to consider other variables such as space in multi-dimensional signals.

### Time Shift

Time shift operation for a continuous time signal replaces the time variable $t$ of $x(t)$ by $t' = (t - T) $ to obtain $x(t') = x (t - T)$, where $T$ is the amount of shift.

Let us see the effect of this operation on a unit step signal, also called as the Heaviside step function (after [Oliver Heaviside](https://en.wikipedia.org/wiki/Oliver_Heaviside)).

```python
import sympy as sp
sp.init_printing()

t = sp.symbols('t', real = True)

#Amount of time shift
T = 2

# The original signal corresponds to $u(t)$ in our lecture notes
x = sp.Heaviside(t)

# New time-shifted signal:
x_T = x.subs(t, t - T) # We can also scaled the signal amplitude by 0.8
                            # so that both x and x_t are visible in the plot. Try it

# Plot
px = sp.plot(x, (t, -3, 3), xlabel = '$t$', label = '$x(t) = u(t)$', legend = True, show = False)
px_T = sp.plot(x_T, (t, -3, 3), xlabel = '$t$', label = '$x_T(t) = x(t-2)$', legend = True,
               show = False, line_color = 'orange')

px.append(px_T[0])
px.show()
```

Note: The `subs(a,b)` operation in SymPy basically replaces all occurences of expression `a` with expression `b`, symbolically. This capability offers us an elegant way of manipulating the time variable of signals.

### Time- Reversal

Time reverse operation of a continuous time signal changes the sign of the time variable $t$ of the signal, $x(t)$, by $t' = -t$ to obtain
$x(t') = x(-t)$. This opertation basically flips the signal along the horizontal axis or reflects the signals with the vertical axis as the mirror. Let us see its effect on a unit step signal:

```python
# Define a unit step signal
x = sp.Heaviside(t - 1)

# Time-reversed signal
x_R = x.subs(t, -t)

# Plot
px = sp.plot(x, (t, -3, 3), xlabel = '$t$', label = '$x(t) = u(t - 1)$', legend = True, show = False)
px_R = sp.plot(x_R, (t, -3, 3), xlabel = '$t$', label = '$x_R(t) = x(-t)$', legend = True,
               show = False, line_color = 'orange')

px.append(px_R[0])
px.show()
```

Let us look at a rectangular pulse as another example:

```python
# Start with a time-shifted rectangular pulse:
x = sp.Heaviside(t - 0.5) - sp.Heaviside(t - 1.5)

# Time-reverse the
x_R = x.subs(t, -t)

# Plot
px = sp.plot(x, (t, -3, 3), xlabel = '$t$', label = '$x(t) = u(t - 0.5) - u(t + 1.5)$',
             legend = True, show = False)
px_R = sp.plot(x_R, (t, -3, 3), xlabel = '$t$', label = '$x_R(t) = x(-t)$', legend = True,
               show = False, line_color = 'orange')

px.append(px_R[0])
px.show()
```

### Time-Scaling

Time scale operation of a continuous time function replaces the time variable, $t$ of $x(t)$ by $t' = at$ to obtain $x(t') = x(at)$, where $a$ is the time scale parameter.

Let us apply it to a periodic continuous time signal:

```python
# Time-scaling parameter:

a = 2.5

x = sp.cos(t)
x_a = x.subs(t, a*t)

# Plot
px = sp.plot(x, (t, -10, 10), xlabel = '$t$', label = '$x(t) = cos(t)$', legend = True, show = False)
px_a = sp.plot(x_a, (t, -10, 10), xlabel = '$t$', label = '$x_a(t) = cos(at)$', legend = True,
               show = False, line_color = 'orange')

px.append(px_a[0])
px.show()
```

Play with the value of `a` above. As you will see, values larger than
 $1$ squeezes the signal, whereas values less than $1$ (and greater than
) stretches the signal.

Also,note that the time reverse operation is a special case of time scaling where $a = -1$.

Let us look at another example below. Here we will apply time scaling to a rectangular pulse.



```python
# Rectangular Pulse
x = sp.Heaviside(t + 0.5) - sp.Heaviside(t - 0.5)

# Applying Time-scaling parameter:

a = 3
x_a = 0.7 * x.subs(t, a*t) # We scle the amplitude to see it better.

# Another one:
b = 1/3
x_b = 0.9 * x.subs(t, b*t)

# Plot
px = sp.plot(x, (t, -3, 3), xlabel = '$t$', label = '$x(t) = u(t + 0.5) - u(t - 0.5)$',
             legend= True, show = False)
px_a = sp.plot(x_a, (t, -3, 3), xlabel = '$t$', label = '$x_a(t) = 0.7x(3t)$',
               legend= True, show = False, line_color = 'orange')
px_b = sp.plot(x_b, (t, -3, 3), xlabel = '$t$', label = '$x_b(t) = 0.9x(t/3)$',
               legend= True, show = False, line_color = 'green')

px.extend(px_a)
px.extend(px_b)
px.show()
```

### Time scale and shift

We can apply time-scaling and time-shifting operations simultaneously to obtain $x(at + b)$ from $x(t)$.

See the example below:

```python
# Define a rectangular pulse:
rect = sp.Heaviside(t + 0.5) - sp.Heaviside(t - 0.5)

# then create another signal which is a combination of two rectangular signals:
x = rect + 2 * rect.subs(t, t -1)

# time scale and shift applied together:
a = 4
b = 5
x_ab = x.subs(t, a*t + b)

# plot
px = sp.plot(x, (t, -3, 3), xlabel = '$t$', label = '$x(t) = u(t + 0.5) - u(t- 0.5)',
             legend = True, show = False)
px_ab = sp.plot(x_ab, (t, -3, 3), xlabel = '$t$', label = '$x_{ab}(t) = x(4t + 5)',
                legend = True, show = False, line_color = 'orange')

px.extend(px_ab)
px.show()
```

Notice how the signal $x(t)$ gets squeezed and shifted at the same time. Play with the values `a` and `b` and see their effect. We especially encourage to try negative values for a to obtain a time reverse effect.

Finally, let us look at an example where we apply time scale and shift operations to a discrete time signal.

In discrete time signals, the time (independent) variable is by definition an integer. And, when we apply time scaling, the time variable might become a non-integer number. For such cases, it is conventional to set the value of the signal to zero. One possible implementation to achieve this is given below.

```python
import numpy as np
from matplotlib import pyplot as plt

# define the time variable.
n = np.arange(-10,10, 1.0)

# define signal x[n]
def x(n):
    # if the time variable is not an integer, return 0
    if not n==np.rint(n):
        print("argument not integer!")
        return 0

    return 1*np.logical_and(n>=-2,n<=2)

# the function above takes a scalar. for it to work on a vector of elements, we
# should vectorize is:
vx = np.vectorize(x)

# apply time scale and shift to x[n] to obtain a new signal
# y[n] = 0.5*x[0.5*(n-2)]
def y(n):
    a = 0.5
    b = -3
    return .5*vx(a*(n+b))

plt.stem(n, vx(n), linefmt='black')
plt.stem(n, y(n), linefmt='orange');

plt.legend((r'$x[n]$', r'$y[n]$'));
```

Here we scaled the value of $y[n]$ by 0.5, so that both signals are visible.

Try different values for `a` and `b` above, which define the new signal $y[n]$.

For $\mid a \mid < 1$, the signal is expanded by inserting zero values in between the time instances of the original signal. This opetation is called **expansion**.

For $\mid a \mid > 1$, the signal is squeezed by discarding the samples of the original signal between $an$ and $a(n -1)$ time instances, for all
$n$. This operation is called **decimation**. You can obtain an example for decimation by setting `a` larger than 1 above.

```python
# apply time scale and shift to x[n] to obtain a new signal
def y(n):
    a = 2
    b = -5
    return .5*vx(a*(n+b))

plt.stem(n, vx(n), linefmt='black')
plt.stem(n, y(n), linefmt='orange');

plt.legend((r'$x[n]$', r'$y[n]$'));
```

## Decomposing a Signal into its Even and Odd Components

Any signal $x(t)$ can be decomposed into even and odd signals as follows.

$$x(t) = \frac{1}{2}x(t) + \frac{1}{2}x(-t) + \frac{1}{2}x(t) - \frac{1}{2}x(-t)$$

where the first two terms on the right hand side

$$ x_e(t) = \frac{1}{2}x(t) + \frac{1}{2}x(-t)$$ is an **even function** and the remaining two terms
$$x_o(t) = \frac{1}{2}x(t) - \frac{1}{2}x(-t)$$ is an **odd function**.

**Task: Give a proof for the above.**

With these definitions, decomposing a given signal into its even and odd parts becomes straightforward as shown below.

```python
import sympy as sym
sym.init_printing()

t = sym.symbols('t', real=True)

def decompose(x):
    # apply time-reverse to x(t):
    xrev = x.subs(t, -t)

    xe = 0.5*x + 0.5*xrev
    xo = 0.5*x - 0.5*xrev

    return xe, xo
```

Let us now apply this decomposition to the following signal: $$ x(t) = \cos( t + \frac{\pi}{3}).$$

```python
# the signal
x = sym.cos(t+sym.pi/3)

# decompose it
xe, xo = decompose(x)

# plot the original signal
px = sym.plot(x, (t, -10, 10), legend=True, label='$x(t)=cos(t+\pi/3)$',
              show=False);
px.show()

# on a separate plot, show even and odd components
px = sym.plot(xe, (t, -10, 10), legend=True, label='$x_e(t)$',
              show=False, line_color='green');
py = sym.plot(xo, (t, -10, 10), legend=True, label=r'$x_o(t)$', show=False,
              line_color='orange')
py.extend(px)
py.show()
```
